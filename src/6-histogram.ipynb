### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python

import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
bins = 5  # Ensure this is a valid integer greater than 0

# Error handling: Validate bins parameter
assert isinstance(bins, int) and bins > 0, "Bins must be a positive integer"

# Create the histogram
plt.hist(data, bins=bins, color='blue', edgecolor='black')

# Customize the plot
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the histogram
plt.show()
```
